```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate counts
counts = df.groupby("SeniorCitizen")["Churn"].value_counts().unstack(fill_value=0)

# Convert to percentage
percent = counts.div(counts.sum(axis=1), axis=0) * 100

# Plot
plt.figure(figsize=(5,5))
bottom = 0

for col in percent.columns:
    plt.bar(percent.index, percent[col], bottom=bottom, label=col)
    bottom += percent[col]

# Add % labels on bars
for i in range(len(percent)):
    y_offset = 0
    for col in percent.columns:
        value = percent.iloc[i][col]
        plt.text(i, y_offset + value/2, f"{value:.1f}%", ha='center', va='center', fontsize=10)
        y_offset += value

plt.title("Churn % by SeniorCitizen")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```
